### Задание 7

На вертикальной столбиковой диаграмме для самого популярного (popularity) и самого непопулярного жанров покажите средние значения по всем характеристикам. Отрицательные и положительные значения характеристик показывайте разными цветами.

```python
# PUT YOUR CODE HERE
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
popular = data.groupby('music_genre').mean().sort_values(by='popularity').tail(1)
unpopular=data.groupby('music_genre').mean().sort_values(by='popularity').head(1)

charic=['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence']

fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

most_popu=popular[charic].values.tolist()[0]
most_unpopu=unpopular[charic].values.tolist()[0]

colors = ['red' if val < 0 else 'black' for val in most_popu + most_unpopu]
ax1 = axes[0]
ax2 = axes[1]

# ax1.set_title('Popular Genre')
# ax2.set_title('Unpopular Genre')
ax1.set_title('Most Popular Genre'+' '+ popular.index[0])
ax1.set_ylabel('Среднее')
ax1.set_xticklabels(charic, rotation=10)
ax2.set_title('Most Unpopular Genre'+' '+ unpopular.index[0])
ax2.set_ylabel('Среднее')
ax2.set_xticklabels(charic,rotation=7)
ax1.set_ylim([-1,1])
ax2.set_ylim([-1,1])

ax1.bar(charic,most_popu,color=colors)
ax2.bar(charic,most_unpopu,color=colors)


# отображаем графики
plt.show()
```

### Задание 8

1. Добавьте в данные категориальный столбец `длительность_трека`, в котором будут значения: "короткая" (<=3min), "средняя" (>3min, <=5min), "длинная" (>5min).
2. Покажите график плотности распределения данных в столбце `loudness`, сгруппированных по категориям длительности.
3. На круговой диаграмме вида "пончик" покажите численное соотношение треков разной длительности.

```python
import seaborn as sns
import matplotlib.pyplot as plt
data['длительность_трека']=''
for row in data.itertuples():
  if row[7]/60000 <= 3:
    data.at[row[0], 'длительность_трека'] = 'короткая'
  elif row[7]/60000 <= 5:
    data.at[row[0], 'длительность_трека'] = 'средняя'
  else:
    data.at[row[0], 'длительность_трека'] = 'длинная'

sns.kdeplot(data=data,x='loudness',hue='длительность_трека')
plt.show()


duration_categories = data['длительность_трека'].value_counts()
labels = duration_categories.index.tolist()
sizes = duration_categories.values.tolist()
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=duration_categories)
ax1.legend(labels, loc='lower right')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig1.gca().add_artist(centre_circle)

plt.show()
```

```python
import sqlite3

# создаем таблицу в БД
con = sqlite3.connect(":memory:")
cur = con.cursor()
cur.execute("CREATE TABLE music (artist, track, genre);")

# переносим в таблицу данные
cur.executemany("INSERT INTO music (artist, track, genre) "
                "VALUES (?, ?, ?);",
                df[["artist_name", "track_name", "music_genre"]].values.tolist())

# сохраняем изменения
con.commit()

# создаем и выполняем запрос
get_songs_per_artist = "SELECT artist, " \
       "COUNT(artist) AS number_of_songs " \
       "FROM music " \
       "GROUP BY artist " \
       "ORDER BY number_of_songs " \
       "DESC;"
cur.execute(get_songs_per_artist)

result = cur.fetchall()
result[:20]
```
